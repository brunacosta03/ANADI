```python
from sklearn.preprocessing import MinMaxScaler

# Converter atributos categóricos em numéricos
data = pd.get_dummies(df.drop(columns=['Unnamed: 0']))

# Normalizar os dados
scaler = MinMaxScaler()
data_norm = scaler.fit_transform(data[data.columns])
data_norm = pd.DataFrame(data_norm, columns=data.columns)

# Imprimir normalização dos atributos
print("Normalização dos atributos:")
data_norm.head()
```

## Exercise 6
### a)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Apresentar a função linear resultante
print("\n\nModelo de regressão linear simples para a Idade")
X = data_norm[['Idade']].to_numpy()
y = data_norm['IMC']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2,
                                                    random_state = 100)

print("X_train size:", X_train.shape)
print("y_train size:", y_train.shape)
print("X_test size:", X_test.shape)
print("y_test size:", y_test.shape)

# Inicia algoritmo regressao linear
regr = LinearRegression()

# Fit aos dados
regr.fit(X_train,y_train)

print("Eq. da reta: y=", regr.intercept_, "+", regr.coef_, "x")
```

## b)

```python
# Visualizar a reta correspondente e respetivo diagrama de dispersão
sns.regplot(x="Idade", y="IMC", data=data_norm, order=1, ci=None, scatter_kws={'color':'r', 's':9})
plt.show()
```

## c)

```python
import numpy as np

# Fazer previsões nos dados de teste
y_pred = regr.predict(X_test)

# Calcular o erro medio absoluto (MAE)
MAE = np.mean(abs(y_test - y_pred))
print("MAE on training set: {:.3f}".format(MAE))

# Calcular a raiz quadrada do erro médio (RMSE
RMSE = np.sqrt(np.mean((y_test-y_pred)**2))
print("RMSE: {:.3f}".format(RMSE))
```

## d)

```python
# Modelo de regressão linear simples para o Peso
print("\n\nModelo de regressão linear simples para o Peso")
X_peso = data_norm[['Peso']].to_numpy()

X_train_peso, X_test_peso, y_train_peso, y_test_peso = train_test_split(X_peso, y, train_size=0.8, test_size=0.2,
                                                                        random_state=100)
print("X_train_peso size:", X_train_peso.shape)
print("y_train_peso size:", y_train_peso.shape)
print("X_test_peso size:", X_test_peso.shape)
print("y_test_peso size:", y_test_peso.shape)

# Inicia algoritmo regressao linear para o peso
regr_peso = LinearRegression()

# Fit aos dados
regr_peso.fit(X_train_peso,y_train_peso)

print("Eq. da reta: y=", regr_peso.intercept_, "+", regr_peso.coef_, "x")

# Fazer previsões nos dados de teste
y_pred_peso = regr_peso.predict(X_test_peso)

# Calcular o erro medio absoluto (MAE)
MAE_peso = np.mean(abs(y_test_peso - y_pred_peso))
print("MAE_peso on training set: {:.3f}".format(MAE_peso))

# Calcular a raiz quadrada do erro médio (RMSE)
RMSE_peso = np.sqrt(np.mean((y_test_peso-y_pred_peso)**2))
print("RMSE_peso: {:.3f}".format(RMSE_peso))
```

# Exercise 8

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred_multi_linear = lr_model.predict(X_test_data)  # Regressão linear múltipla
y_pred_tree = dt_regressor.predict(X_test_data)  # Árvore de regressão
y_pred_nn = nn.predict(X_test_data)  # Rede neural


mae_multi_linear = mean_absolute_error(y_test_data, y_pred_multi_linear)
rmse_linear = np.sqrt(mean_squared_error(y_test_data, y_pred_multi_linear))

mae_tree = mean_absolute_error(y_test_data, y_pred_tree)
rmse_tree = np.sqrt(mean_squared_error(y_test_data, y_pred_tree))

mae_nn = mean_absolute_error(y_test_data, y_pred_nn)
rmse_nn = np.sqrt(mean_squared_error(y_test_data, y_pred_nn))

print("Resultados da Regressão Linear Múltipla:")
print("MAE:", round(mae_multi_linear, 3))
print("RMSE:", round(rmse_linear, 3))

print("\nResultados da Árvore de Regressão:")
print("MAE:", round(mae_tree, 3))
print("RMSE:", round(rmse_tree, 3))

print("\nResultados da Rede Neural:")
print("MAE:", round(mae_nn, 3))
print("RMSE:", round(rmse_nn, 3))
```

# Exercise 9

```python
from statsmodels.stats._lilliefors import lilliefors
from scipy.stats import wilcoxon

alpha = 0.05
print("\nalpha: ", alpha)

print("\np-value lilliefors RML: ", round(lilliefors(y_pred_multi_linear)[1], 3))
print("p-value lilliefors Árvore: ", round(lilliefors(y_pred_tree)[1], 3))

pvalue = wilcoxon(y_pred_multi_linear, y_pred_tree, alternative="two-sided").pvalue
print('\nwilcoxon pvalue: ', round(pvalue, 3))

if pvalue < alpha:
    print('As diferenças entre a RML e a Árvore são relevantes.')
else:
    print('As diferenças entre a RML e a Árvore não são relevantes')
```
