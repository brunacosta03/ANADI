```python
import numpy as np
import tensorflow as tf

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, recall_score
from sklearn.tree import plot_tree
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from scipy.stats import ttest_rel

# Change the target variable to integer
df['Label']= np.where(df['Label'] == 'Magreza_Grau_III', 0, df['Label'])
df['Label']= np.where(df['Label'] == 'Magreza_Grau_II', 1, df['Label'])
df['Label']= np.where(df['Label'] == 'Magreza_Grau_I', 2, df['Label'])
df['Label']= np.where(df['Label'] == 'Peso_Normal', 3, df['Label'])
df['Label']= np.where(df['Label'] == 'Excesso_Peso_Grau_I', 4, df['Label'])
df['Label']= np.where(df['Label'] == 'Excesso_Peso_Grau_II', 5, df['Label'])
df['Label']= np.where(df['Label'] == 'Obesidade_Moderada', 6, df['Label'])
df['Label']= np.where(df['Label'] == 'Obesidade_Severa', 7, df['Label'])
df['Label']= np.where(df['Label'] == 'Obesidade_Mórbida', 8, df['Label'])

df['Label'] = df['Label'].astype(int)

x = df[['Idade', 'Altura', 'Peso', 'FCV', 'NRP', 'CA', 'FAF', 'TUDE']]
y = df['Label']
```

### Árvore de Decisão

```python
decisiontree = DecisionTreeClassifier(random_state=100)

grid_parameters = {
    'max_depth': [3, 5, 7, 10, 15, 20, 25, 30, 35, 40, 45, 50,None],
    'min_samples_split': [2, 5, 10,15, 20, 25],
    'min_samples_leaf': [1, 2, 4, 8, 16, 32],
    'criterion': ['gini', 'entropy']
}

# GridSearchCV is used to find the best hyperparameters
grid_search = GridSearchCV(decisiontree, grid_parameters, cv=5, scoring='accuracy')
grid_search.fit(x, y)

best_model = grid_search.best_estimator_

kf = KFold(n_splits=5, shuffle=True, random_state=100)
scores_dt = {'accuracy': [], 'sensitivity': [], 'specificity': [], 'f1': []}

for train_index, test_index in kf.split(x):
    # iloc is used to select rows and columns by number
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    best_model.fit(x_train, y_train)
    y_pred = best_model.predict(x_test)
    
    scores_dt['accuracy'].append(accuracy_score(y_test, y_pred))
    scores_dt['sensitivity'].append(recall_score(y_test, y_pred, average='macro'))
    scores_dt['f1'].append(f1_score(y_test, y_pred, average='macro'))

    matrix = confusion_matrix(y_test, y_pred)
    specs_class = []
    # Specificity is calculated for each class
    for i in range(len(matrix)):
        tn = np.sum(matrix) - np.sum(matrix[i]) - np.sum(matrix[:, i]) + matrix[i, i]
        fp = np.sum(matrix[:, i]) - matrix[i, i]
        specs_class.append(tn / (tn + fp))
    scores_dt['specificity'].append(np.mean(specs_class))
    
dt_mean_accuracy = np.mean(scores_dt['accuracy'])
std_dt_accuracy = np.std(scores_dt['accuracy'])
dt_mean_specificity = np.mean(scores_dt['specificity'])
std_dt_specificity = np.std(scores_dt['specificity'])
dt_mean_sensitivity = np.mean(scores_dt['sensitivity'])
std_dt_sensitivity = np.std(scores_dt['sensitivity'])
dt_mean_f1 = np.mean(scores_dt['f1'])
std_dt_f1 = np.std(scores_dt['f1'])

print("Accuracy: {:.4f} +/- {:.4f}".format(dt_mean_accuracy, std_dt_accuracy))
print("Specificity: {:.4f} +/- {:.4f}".format(dt_mean_specificity, std_dt_specificity))
print("Sensitivity: {:.4f} +/- {:.4f}".format(dt_mean_sensitivity, std_dt_sensitivity))
print("F1 Score: {:.4f} +/- {:.4f}".format(dt_mean_f1, std_dt_f1))
print('Best Parameters: ', grid_search.best_params_)
```

```python
df['Genero_Feminino'] = np.where(df['Genero'] == 'Feminino', 1, 0)

x = df[['Idade', 'Altura', 'Peso', 'FCV','CA', 'FAF', 'TUDE']]
y = df['Genero_Feminino']

labelEncoder = LabelEncoder()
y = labelEncoder.fit_transform(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

# scaler is used to standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train_df = pd.DataFrame(x_train, columns=x.columns)
x_test_df = pd.DataFrame(x_test, columns=x.columns)
y_train_df = pd.Series(y_train)
y_test_df = pd.Series(y_test)

tf.random.set_seed(100)

def create_model():
    model = keras.models.Sequential([
        keras.layers.InputLayer(input_shape=x_train_df.shape[1:]),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(16, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = create_model()

history = model.fit(x_train, y_train, validation_split=0.1, epochs=30, batch_size=32)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Model Accuracy")
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')

loss, accuracy = model.evaluate(x_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)

plt.tight_layout()
plt.show()

kf = KFold(n_splits=5, shuffle=True, random_state=100)
scores = {'accuracy': [], 'sensitivity': [], 'specificity': [], 'f1': []}

for train_index, test_index in kf.split(x_train_df):
    # iloc is used to select rows and columns by number
    x_train_fold, x_test_fold = x_train_df.iloc[train_index], x_train_df.iloc[test_index]
    y_train_fold, y_test_fold = y_train_df.iloc[train_index], y_train_df.iloc[test_index]

    model = create_model()
    model.fit(x_train_fold, y_train_fold, epochs=30, batch_size=32, verbose=0)

    y_pred = np.round(model.predict(x_test_fold))

    scores['accuracy'].append(accuracy_score(y_test_fold, y_pred))
    scores['sensitivity'].append(recall_score(y_test_fold, y_pred, average='macro'))
    scores['f1'].append(f1_score(y_test_fold, y_pred, average='macro'))

    matrix = confusion_matrix(y_test_fold, y_pred)
    specs_class = []
    # Specificity is calculated for each class
    for i in range(len(matrix)):
        tn = np.sum(matrix) - np.sum(matrix[i]) - np.sum(matrix[:, i]) + matrix[i, i]
        fp = np.sum(matrix[:, i]) - matrix[i, i]
        specs_class.append(tn / (tn + fp))
    scores['specificity'].append(np.mean(specs_class))
    
rnn_mean_accuracy = np.mean(scores['accuracy'])
std_rnn_accuracy = np.std(scores['accuracy'])
rnn_mean_specificity = np.mean(scores['specificity'])
std_rnn_specificity = np.std(scores['specificity'])
rnn_mean_sensitivity = np.mean(scores['sensitivity'])
std_rnn_sensitivity = np.std(scores['sensitivity'])
rnn_mean_f1 = np.mean(scores['f1'])
std_f1 = np.std(scores['f1'])

print("Accuracy: {:.4f} +/- {:.4f}".format(rnn_mean_accuracy, std_rnn_accuracy))
print("Specificity: {:.4f} +/- {:.4f}".format(rnn_mean_specificity, std_rnn_specificity))
print("Sensitivity: {:.4f} +/- {:.4f}".format(rnn_mean_sensitivity, std_rnn_sensitivity))
print("F1 Score: {:.4f} +/- {:.4f}".format(rnn_mean_f1, std_f1))
```

```python
x = df[['Idade', 'Altura', 'Peso', 'FCV','CA', 'FAF', 'TUDE']]
y = df['Genero_Feminino']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
    svm = SVC(kernel=kernel)

    kf = KFold(n_splits=5, shuffle=True, random_state=100)
    scores = {'accuracy': [], 'sensitivity': [], 'specificity': [], 'f1': []}

    for train_index, test_index in kf.split(x_train):
        # iloc is used to select rows and columns by number
        x_train_fold, x_test_fold = x_train.iloc[train_index], x_train.iloc[test_index]
        y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

        svm.fit(x_train_fold, y_train_fold)
        y_pred = svm.predict(x_test_fold)

        scores['accuracy'].append(accuracy_score(y_test_fold, y_pred))
        scores['sensitivity'].append(recall_score(y_test_fold, y_pred, average='macro'))
        scores['f1'].append(f1_score(y_test_fold, y_pred, average='macro'))

        matrix = confusion_matrix(y_test_fold, y_pred)
        specs_class = []
        # Specificity is calculated for each class
        for i in range(len(matrix)):
            tn = np.sum(matrix) - np.sum(matrix[i]) - np.sum(matrix[:, i]) + matrix[i, i]
            fp = np.sum(matrix[:, i]) - matrix[i, i]
            specs_class.append(tn / (tn + fp))
        scores['specificity'].append(np.mean(specs_class))

    svm_mean_accuracy = np.mean(scores['accuracy'])
    std_svm_accuracy = np.std(scores['accuracy'])
    svm_mean_specificity = np.mean(scores['specificity'])
    std_svm_specificity = np.std(scores['specificity'])
    svm_mean_sensitivity = np.mean(scores['sensitivity'])
    std_svm_sensitivity = np.std(scores['sensitivity'])
    svm_mean_f1 = np.mean(scores['f1'])
    std_svm_f1 = np.std(scores['f1'])

    print("\nKERNEL TYPE: ", kernel)
    print("Accuracy: {:.4f} +/- {:.4f}".format(svm_mean_accuracy, std_svm_accuracy))
    print("Specificity: {:.4f} +/- {:.4f}".format(svm_mean_specificity, std_svm_specificity))
    print("Sensitivity: {:.4f} +/- {:.4f}".format(svm_mean_sensitivity, std_svm_sensitivity))
    print("F1 Score: {:.4f} +/- {:.4f}".format(svm_mean_f1, std_svm_f1))
    
    # Save the best model
    if kernel == 'linear':
        best_svm_accuracy = svm_mean_accuracy
        best_svm_specificity = svm_mean_specificity
        best_svm_sensitivity = svm_mean_sensitivity
        best_svm_f1 = svm_mean_f1
```
